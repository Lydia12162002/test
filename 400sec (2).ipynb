```python
# =========================================================
# 0) 安裝套件（第一次跑需要）
# =========================================================
try:
    import mlxtend
except ImportError:
    !pip -q install mlxtend scikit-learn openpyxl

# =========================================================
# 1) 匯入套件 & 基本設定
# =========================================================
import os
import datetime
import time
import itertools
from collections import Counter

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import MultiLabelBinarizer
from mlxtend.frequent_patterns import apriori, fpgrowth, association_rules

try:
    from IPython.display import display
except Exception:
    pass

DATA_PATH = "交易資料集(2).xlsx"
USE_ITEM_COL = "ITEM_NO"     # 或 "ITEM_ID"
INVOICE_COL = "INVOICE_NO"
QTY_COL = "QUANTITY"

# ✅ 你要比 2 / 3 就改這裡：
MAX_ITEMSET_LEN = 3          # <- 改成 2 或 3

# Apriori 防爆
APRIORI_LOW_MEMORY = True

# ✅ FP-Growth 的 OHE 轉 dense（避免 sparse 造成額外慢）
FP_DENSE_MAX_COLS = 5000
FP_DENSE_MAX_CELLS = 50_000_000

pd.set_option("display.max_colwidth", 120)

# =========================================================
# ✅【新增】把表格/圖片存到「Jupyter 資料夾」（目前工作目錄）
# =========================================================
EXPORT_DIR = os.getcwd()
print("✅ 目前工作目錄（Jupyter 資料夾）:", EXPORT_DIR)

def _ts():
    return datetime.datetime.now().strftime("%Y%m%d_%H%M%S")

def save_df(df: pd.DataFrame, basename_no_ext: str):
    """輸出 CSV + XLSX 到 Jupyter 資料夾"""
    csv_path  = os.path.join(EXPORT_DIR, f"{basename_no_ext}.csv")
    xlsx_path = os.path.join(EXPORT_DIR, f"{basename_no_ext}.xlsx")
    df.to_csv(csv_path, index=False, encoding="utf-8-sig")
    df.to_excel(xlsx_path, index=False)
    print("✅ 表格已存到:", csv_path)
    print("✅ 表格已存到:", xlsx_path)

def save_plot(fig, basename_no_ext: str, dpi=300):
    """輸出 PNG 到 Jupyter 資料夾"""
    png_path = os.path.join(EXPORT_DIR, f"{basename_no_ext}.png")
    fig.savefig(png_path, dpi=dpi, bbox_inches="tight")
    print("✅ 圖片已存到:", png_path)


# =========================================================
# 2) 讀取資料 + 前處理（剔除 QUANTITY<=0）
# =========================================================
def load_and_preprocess(path: str,
                        invoice_col: str = "INVOICE_NO",
                        item_col: str = "ITEM_NO",
                        qty_col: str = "QUANTITY"):
    df = pd.read_excel(path)

    need_cols = {invoice_col, item_col, qty_col}
    miss = need_cols - set(df.columns)
    if miss:
        raise ValueError(f"缺少必要欄位：{miss}，目前欄位：{list(df.columns)}")

    # 剔除退貨/註銷（數量為 0 或負）
    df = df[df[qty_col] > 0].copy()

    # 品項轉字串
    df[item_col] = df[item_col].astype(str).str.strip()

    # 交易清單（每張發票的品項去重）
    tx = (
        df.groupby(invoice_col)[item_col]
          .apply(lambda s: list(pd.unique(s)))
          .tolist()
    )
    return df, tx


df, transactions = load_and_preprocess(DATA_PATH, INVOICE_COL, USE_ITEM_COL, QTY_COL)
print("前處理後資料列數:", len(df))
print("交易(發票)筆數:", len(transactions))
print("平均每筆交易品項數:", np.mean([len(t) for t in transactions]))
print("是否仍有 QUANTITY<=0 :", (df[QTY_COL] <= 0).any())


# =========================================================
# 3) 支持度過濾 + OHE 建立
# =========================================================
def filter_items_by_support(transactions, min_support: float):
    n_tx = len(transactions)
    cnt = Counter()
    for t in transactions:
        cnt.update(set(t))

    kept = {item for item, c in cnt.items() if (c / n_tx) >= min_support}
    filtered_tx = [[i for i in set(t) if i in kept] for t in transactions]
    filtered_tx = [t for t in filtered_tx if len(t) > 0]
    return filtered_tx, kept


def make_ohe_sparse_for_apriori(transactions, min_support: float):
    """Apriori 用：pandas sparse DataFrame（省 RAM）"""
    tx_f, _ = filter_items_by_support(transactions, min_support=min_support)

    mlb = MultiLabelBinarizer(sparse_output=True)
    X = mlb.fit_transform(tx_f)

    ohe = pd.DataFrame.sparse.from_spmatrix(X, columns=mlb.classes_).astype(bool)
    return ohe, tx_f, "sparse"


def make_ohe_for_fpgrowth(transactions, min_support: float):
    """FP-Growth 用：能 dense 就 dense（通常更快）"""
    tx_f, _ = filter_items_by_support(transactions, min_support=min_support)

    mlb = MultiLabelBinarizer(sparse_output=True)
    X = mlb.fit_transform(tx_f)

    n_rows, n_cols = X.shape
    cells = n_rows * n_cols

    if (n_cols <= FP_DENSE_MAX_COLS) and (cells <= FP_DENSE_MAX_CELLS):
        ohe = pd.DataFrame(X.toarray().astype(bool), columns=mlb.classes_)
        kind = "dense"
    else:
        ohe = pd.DataFrame.sparse.from_spmatrix(X, columns=mlb.classes_).astype(bool)
        kind = "sparse"

    return ohe, tx_f, kind


# =========================================================
# 4) 冗餘規則刪除（照題目定義）
# 若存在 X' ⊂ X 且 conf(X'→Y) > conf(X→Y)，則 X→Y 為冗餘
# =========================================================
def prune_redundant_rules_conf(rules_df: pd.DataFrame) -> pd.DataFrame:
    if rules_df.empty:
        return rules_df

    redundant_idx = set()

    for cons, grp in rules_df.groupby("consequents"):
        conf_map = {row["antecedents"]: row["confidence"] for _, row in grp.iterrows()}

        for idx, row in grp.iterrows():
            A = row["antecedents"]
            conf_A = row["confidence"]
            items = list(A)
            if len(items) <= 1:
                continue

            found = False
            for r in range(1, len(items)):
                for comb in itertools.combinations(items, r):
                    A_sub = frozenset(comb)
                    if A_sub in conf_map and conf_map[A_sub] > conf_A:
                        found = True
                        break
                if found:
                    break

            if found:
                redundant_idx.add(idx)

    return rules_df.drop(index=list(redundant_idx)).reset_index(drop=True)


# =========================================================
# 5) 建 itemsets（✅同一 support 只算一次）
# =========================================================
def build_itemsets_once(transactions, algorithm: str, min_support: float, max_len: int = MAX_ITEMSET_LEN):
    algo = algorithm.lower().strip()
    if algo not in {"apriori", "fpgrowth"}:
        raise ValueError("algorithm 必須是 'apriori' 或 'fpgrowth'")

    t0 = time.perf_counter()

    # OHE
    if algo == "fpgrowth":
        ohe, tx_filtered, ohe_kind = make_ohe_for_fpgrowth(transactions, min_support=min_support)
    else:
        ohe, tx_filtered, ohe_kind = make_ohe_sparse_for_apriori(transactions, min_support=min_support)

    t1 = time.perf_counter()

    # frequent itemsets
    if algo == "apriori":
        fi = apriori(
            ohe,
            min_support=min_support,
            use_colnames=True,
            low_memory=APRIORI_LOW_MEMORY,
            max_len=max_len
        )
    else:
        fi = fpgrowth(
            ohe,
            min_support=min_support,
            use_colnames=True,
            max_len=max_len
        )

    t2 = time.perf_counter()

    base_info = {
        "algorithm": algo,
        "min_support": min_support,
        "max_len": max_len,
        "ohe_kind": ohe_kind,
        "n_transactions_before": len(transactions),
        "n_transactions_after_filter": len(tx_filtered),
        "n_items_after_filter": ohe.shape[1],
        "n_frequent_itemsets": len(fi),
        "time_ohe_sec": (t1 - t0),
        "time_itemsets_sec": (t2 - t1),
        "time_ohe_plus_itemsets_sec": (t2 - t0),
    }
    return fi, base_info


# =========================================================
# 6) 由 itemsets 產生規則（confidence 變動只重跑這段）
# =========================================================
def rules_from_itemsets(fi: pd.DataFrame,
                        min_confidence: float,
                        prune_redundant: bool = True,
                        single_item_consequent: bool = True):
    t0 = time.perf_counter()

    if fi is None or fi.empty:
        rules = pd.DataFrame(columns=["antecedents","consequents","support","confidence","lift"])
    else:
        rules = association_rules(fi, metric="confidence", min_threshold=min_confidence)

    if not rules.empty and single_item_consequent:
        rules = rules[rules["consequents"].apply(lambda s: len(s) == 1)].copy()

    if not rules.empty and prune_redundant:
        rules = prune_redundant_rules_conf(rules)

    if not rules.empty:
        rules = rules.sort_values(["confidence", "lift", "support"], ascending=False).reset_index(drop=True)

    t1 = time.perf_counter()
    return rules, (t1 - t0)


# =========================================================
# 7) mine_rules：一次跑完（單一組參數用）
# =========================================================
def mine_rules(transactions,
               algorithm: str = "apriori",
               min_support: float = 0.001,
               min_confidence: float = 0.5,
               prune_redundant: bool = True,
               single_item_consequent: bool = True,
               max_len: int = MAX_ITEMSET_LEN):

    fi, base = build_itemsets_once(transactions, algorithm, min_support, max_len=max_len)
    rules, time_rules = rules_from_itemsets(
        fi, min_confidence,
        prune_redundant=prune_redundant,
        single_item_consequent=single_item_consequent
    )

    info = {
        **base,
        "min_confidence": min_confidence,
        "n_rules": len(rules),
        "time_rules_sec": time_rules,
        "time_total_sec": base["time_ohe_plus_itemsets_sec"] + time_rules,
    }
    return rules, info


# =========================================================
# 8) 規則存檔/讀檔（CSV）
# =========================================================
def _set_to_str(s: frozenset) -> str:
    return "|".join(sorted(list(s)))

def _str_to_set(x: str) -> frozenset:
    x = (x or "").strip()
    if x == "":
        return frozenset()
    return frozenset(x.split("|"))

def save_rules_csv(rules_df: pd.DataFrame, path: str):
    out = rules_df.copy()
    out["antecedents"] = out["antecedents"].apply(_set_to_str)
    out["consequents"] = out["consequents"].apply(_set_to_str)
    out.to_csv(path, index=False, encoding="utf-8-sig")

def load_rules_csv(path: str) -> pd.DataFrame:
    df = pd.read_csv(path)
    df["antecedents"] = df["antecedents"].apply(_str_to_set)
    df["consequents"] = df["consequents"].apply(_str_to_set)
    return df


# =========================================================
# 9) 推薦功能：同次不重複 + 不推薦已在籃子內
# =========================================================
def recommend_products(input_items, rules_df: pd.DataFrame, top_k: int = 10):
    if rules_df.empty:
        return []

    basket = set(map(str, input_items))
    cand = rules_df[rules_df["antecedents"].apply(lambda A: A.issubset(basket))].copy()
    if cand.empty:
        return []

    best = {}
    for _, r in cand.iterrows():
        rhs_item = next(iter(r["consequents"]))
        if rhs_item in basket:
            continue

        score = (r["confidence"], r["lift"], r["support"])
        if (rhs_item not in best) or (score > best[rhs_item]["score"]):
            best[rhs_item] = {
                "item": rhs_item,
                "score": score,
                "confidence": r["confidence"],
                "lift": r["lift"],
                "support": r["support"],
                "antecedents": r["antecedents"],
            }

    ranked = sorted(best.values(), key=lambda d: d["score"], reverse=True)
    return ranked[:top_k]


# =========================================================
# 10) ✅ 參數掃描（改完：同一個 support 只算一次 itemsets）
# =========================================================
def parameter_sweep(transactions, supports, confidences, algorithm: str, max_len: int = MAX_ITEMSET_LEN):
    rows = []
    for s in supports:
        fi, base = build_itemsets_once(transactions, algorithm, s, max_len=max_len)

        for c in confidences:
            rules, time_rules = rules_from_itemsets(
                fi, c,
                prune_redundant=True,
                single_item_consequent=True
            )

            n_unique_rhs = rules["consequents"].apply(lambda x: next(iter(x))).nunique() if not rules.empty else 0

            info = {
                **base,
                "min_confidence": c,
                "n_rules": len(rules),
                "n_unique_recommendable_items": n_unique_rhs,
                "time_rules_sec": time_rules,
                "time_total_sec": base["time_ohe_plus_itemsets_sec"] + time_rules,
            }
            rows.append(info)

            print(f"[{algorithm}] support={s:.6f}, conf={c:.2f} => "
                  f"rules={info['n_rules']}, rhs_unique={n_unique_rhs}, "
                  f"ohe={base['ohe_kind']}, time≈{info['time_total_sec']:.2f}s")

    return pd.DataFrame(rows)


# =========================================================
# 11) 實際執行：比較 Apriori vs FP-Growth（已加速 sweep）
# =========================================================
supports = [0.0020, 0.0015, 0.0010, 0.0008]
confidences = [0.3, 0.5, 0.7]

tA0 = time.perf_counter()
result_apriori  = parameter_sweep(transactions, supports, confidences, algorithm="apriori",  max_len=MAX_ITEMSET_LEN)
tA1 = time.perf_counter()

tF0 = time.perf_counter()
result_fpgrowth = parameter_sweep(transactions, supports, confidences, algorithm="fpgrowth", max_len=MAX_ITEMSET_LEN)
tF1 = time.perf_counter()

print(f"\n➡ Apriori   sweep總時間: {(tA1 - tA0):.2f} 秒")
print(f"➡ FP-Growth sweep總時間: {(tF1 - tF0):.2f} 秒")

display(result_apriori.sort_values(["min_support","min_confidence"]))
display(result_fpgrowth.sort_values(["min_support","min_confidence"]))

# ✅【新增】把兩張結果表存到 Jupyter 資料夾（CSV + XLSX）
save_df(result_apriori.sort_values(["min_support","min_confidence"]), f"result_apriori_{_ts()}")
save_df(result_fpgrowth.sort_values(["min_support","min_confidence"]), f"result_fpgrowth_{_ts()}")


# =========================================================
# 12) 視覺化（✅新增：每張圖都存到 Jupyter 資料夾）
# =========================================================
def plot_param_effect(df_res: pd.DataFrame, title_prefix: str):
    for metric, ylab in [
        ("n_rules", "Number of Rules"),
        ("n_unique_recommendable_items", "Unique RHS Items (Recommendable)"),
        ("time_total_sec", "Total Time (sec)"),
    ]:
        fig = plt.figure()
        for conf in sorted(df_res["min_confidence"].unique()):
            sub = df_res[df_res["min_confidence"] == conf].sort_values("min_support")
            plt.plot(sub["min_support"], sub[metric], marker="o", label=f"conf={conf}")
        plt.xlabel("min_support")
        plt.ylabel(ylab)
        plt.title(f"{title_prefix} - {metric} (max_len={MAX_ITEMSET_LEN})")
        plt.legend()

        # ✅ 存圖（不影響原本流程）
        safe_title = title_prefix.replace(" ", "_").replace("+", "plus").replace("/", "_").replace("(", "").replace(")", "")
        save_plot(fig, f"{safe_title}_{metric}_maxlen{MAX_ITEMSET_LEN}_{_ts()}")

        plt.show()

plot_param_effect(result_apriori,  "Apriori")
plot_param_effect(result_fpgrowth, f"FP-Growth (dense+reuse itemsets)")


# =========================================================
# 13) 產出一組最終規則（示範：FP-Growth, support=0.001, conf=0.5）
# =========================================================
best_rules, best_info = mine_rules(
    transactions,
    algorithm="fpgrowth",
    min_support=0.0010,
    min_confidence=0.5,
    prune_redundant=True,
    single_item_consequent=True,
    max_len=MAX_ITEMSET_LEN
)
print("最終規則設定：", best_info)
display(best_rules.head(20))

RULES_PATH = f"mined_rules_fpgrowth_maxlen{MAX_ITEMSET_LEN}_s001_c05.csv"
save_rules_csv(best_rules, RULES_PATH)
print("已存檔：", RULES_PATH)

loaded_rules = load_rules_csv(RULES_PATH)
print("讀回規則筆數：", len(loaded_rules))

# ✅【新增】把 best_rules 也存成表格（CSV + XLSX）到 Jupyter 資料夾
save_df(best_rules, f"best_rules_{_ts()}")


# =========================================================
# 14) 推薦示範
# =========================================================
sample_basket = next((t for t in transactions if len(set(t)) >= 2), None)
print("範例輸入品項:", sample_basket)

recs = recommend_products(sample_basket, loaded_rules, top_k=10)
print("\n推薦結果（同次不重複、且不推薦已在籃子內）：")
for i, r in enumerate(recs, 1):
    print(f"{i:02d}. 推薦品項={r['item']} | conf={r['confidence']:.3f}, "
          f"lift={r['lift']:.3f}, support={r['support']:.4f} | 規則 {set(r['antecedents'])} -> {r['item']}")
```
